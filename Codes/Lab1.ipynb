# Using while loop

```python
a=0.5
n=1
while 1+a>1:
    a/=2.0
    n+=1
    
print(n-1,'\t',2**(1-n))
```

```python
i=[0.1,0.01,0.001,0.0001,0.00001,0.000001,0.00000001,0.0000000001,0.000000000001,0.000000000000001]
new=[]
def myfun1(x):
    return ((x**2)+1.0)**(0.5)-1.0
def myfun2(x):
    return x**2/(((x**2)+1)**(0.5)+1.0)
###original=[]
for j in i:
    A=((j**2)+1.0)**(0.5)-1.0
    B=j**2/(((j**2)+1)**(0.5)+1.0)
    new.append(abs((A-B)/B))
print(new)
plt.xscale('log')
plt.yscale('log')
plt.plot(i,new)
# plt.legend()
```
